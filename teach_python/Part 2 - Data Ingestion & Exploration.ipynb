## <a name='2_1'>2.1 Pandas Built-in Function for data exploration</a>
### 2.1a df.info()
This method provides a quick overview of the data in terms of data structure, missing values, etc.
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.info.html

```python
print('dataframe shape (row x column):', csv.shape)
print('datatype of the columns:',csv.dtypes)
print('a list of column names',csv.columns)
csv.info()
```

### 2.1b df.describe()
This method generates a descriptive analysis of the columns with int64 or float64 datatype by default. However, can also provide additional parameters (see link below) or if dataframe does not consist of numeric columns to get descriptive analysis of object or non-numeric columns as well.<br>
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html

```python
csv.describe()
```

### 2.1c Series.unique() or Series.nunique()
<b>df['colName'].unique():</b> returns a list of unique values from a column<br>
https://pandas.pydata.org/docs/reference/api/pandas.Series.unique.html

<b>df['colName'].nunique():</b> returns a count of unique values from a column<br>
https://pandas.pydata.org/docs/reference/api/pandas.Series.nunique.html


```python
csv['FinancialClass'].unique()
```

```python
csv['FinancialClass'].nunique()
```

### 2.1d df.nunique()
This method returns a Series object that shows the count of unique values in each column of a dataframe.<br>
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.nunique.html

```python
csv.nunique()
```

### 2.1e Series.value_counts()
<b>df['colName'].value_counts():</b> returns a Series containing counts of unique values within a column.<br>
https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html

```python
csv['FinancialClass'].value_counts()
```

## <a name='2_2'>2.2 Custom Function for data exploration</a>

```python
%%time
def print_full(x):  # Display all df columns and rows
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    display(x)
    pd.reset_option('display.max_rows')
    pd.reset_option('display.max_columns')

# this function is to display the full dataframe. 
# Caution: large dataset that's both wide and long can cause the notebook to crash!
print_full(csv.iloc[:50,:])
```

```python
%%time    
def df_unique(df): #retrieve unique value from each column of a dataframe
    # https://stackoverflow.com/a/68813683
    uni = pd.Series({col:list(df[col].unique()) for col in df})
    return uni

# this function return a Series object that show the unique values in each column of a dataframe.
df_unique(csv)
```

```python
df_uniq = df_unique(csv)
df_uniq['FinancialClass']  #this syntax show the unique values in FinancialClassDesc column
```
